```python
# 라이브러리 호출

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import sklearn
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/example_cluster.csv')
df
```

```python
# 데이터를 산점도로 최초 표현
plt.figure(figsize=(9,7))
sns.scatterplot(x='var_1', y='var_2', data = df)
```

```python
# K평균 군집화 호출

from sklearn.cluster import KMeans
```

```python
kmeans_model = KMeans(n_clusters=3, random_state = 27)
kmeans_model.fit(df)
```

```python
# 예측값 제출

pred = kmeans_model.predict(df)
pred
```

```python
# label 정의

df['label'] = pred
```

```python
plt.figure(figsize=(9,7))
sns.scatterplot(x='var_1', y='var_2', data = df,
                hue = 'label')
```

```python
plt.figure(figsize=(9,7))
sns.scatterplot(x='var_1', y='var_2', data = df,
                hue = 'label', palette='rainbow')
```

```python
kmeans_model.inertia_
```

```python
# k가 500개 이상 다수일경우
kmeans_model2 = KMeans(n_clusters = 500, random_state = 27)
kmeans_model2.fit(df)
```

```python
# inertia_
kmeans_model2.inertia_
```

```python
df['label2'] = kmeans_model2.predict(df)
```

```python
df['label2']
```

```python
plt.figure(figsize=(9,7))
sns.scatterplot(
    x = 'var_1', y='var_2', data = df, hue = 'label2')
```

```python
# elbow method 사용

df = df.drop(['label','label2'],axis = 1)
df.columns
```

```python
df = df.drop(['lael'], axis = 1)
```

```python
# k로 반복함수 지정

distance = []
for k in range(2,10):
    k_model = KMeans(n_clusters = k)
    k_model.fit(df)
    distance.append(k_model.inertia_)
```

```python
distance
```

```python
# elbow 지점 확인
sns.lineplot(x = range(2,10), y = distance)
```

```python
fraud.cc_num.unique()
```

```python
customer = pd.get_dummies(fraud, columns = ['category'])
customer.info()
```

```python
customer
```

```python
customer.iloc[:,2:]
```

```python
# 특정 카테고리 포함여부 확인

for col_name in customer.iloc[:,2:].columns:
    customer[col_name] = customer[col_name] * customer['amt']
```

```python
customer
```

```python
customer_agg = customer.groupby('cc_num').sum()
customer_agg
```

```python
# scaling 시작

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaled = scaler.fit_transform(customer_agg)
scaled
```

```python
scaled_df = pd.DataFrame(scaled, columns = customer_agg.columns,
                        index = customer_agg.index)
scaled_df
```

```python
distance = []
for k in range(2, 10):
    k_model = KMeans(n_clusters = k, random_state = 27)
    k_model.fit(scaled_df)
    distance.append(k_model.inertia_)
```

```python
sns.lineplot(x = range(2,10), y = distance)
```

```python
kmeans_model3 = KMeans(n_clusters = 50, random_state = 27)
kmeans_model3.fit(scaled_df)
```

```python
pred = kmeans_model3.predict(scaled_df)
pred
```

```python
scaled_df['label3'] = pred
```

```python
# 실루엣 계수 사용
from sklearn.metrics import silhouette_score
```

```python
# silhouette 리스트에 클러스터 간 거리를 저장시킴
silhouette = []
for k in range(2,10):
    k_model = KMeans(n_clusters = k, random_state = 27)
    labels = k_model.fit_predict(scaled_df)
    silhouette.append(silhouette_score(scaled_df, labels))
```

```python
sns.lineplot(x = range(2,10), y = silhouette)
```

```python
k_model = KMeans(n_clusters = 4, random_state = 27)
labels = k_model.fit_predict(scaled_df)
```

```python
scaled_df['label'] = labels
scaled_df
```

```python
scaled_df_mean = scaled_df.groupby('label').mean()
scaled_df_count = scaled_df.groupby('label').count().amt
```

```python
scaled_df_count = scaled_df_count.rename('count')
scaled_df_all = scaled_df_mean.join(scaled_df_count)
```

```python
scaled_df_all.sum().sort_values(ascending = False)
```
