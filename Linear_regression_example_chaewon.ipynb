```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# https://www.kaggle.com/
# https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/insurance.csv
df = pd.read_csv('https://raw.githubusercontent.com/bigdata-young/ai_26th/main/data/insurance.csv')
df
```

```python
df.info
pd.options.display.float_format = "{:.2f}".format
```

```python
df.describe()
```

```python
plt.figure(figsize=(12,9))
sns.kdeplot(data=df, x='expenses', hue = 'age', legend = False, palette = 'flare')
plt.title("The younger is cheaper?, see that thing")
```

```python
plt.figure(figsize=(12, 9))
sns.lineplot(x = "smoker", y = "expenses",  hue = 'age', markers = True, legend = False, palette = 'deep', data = df)
plt.title("The smoker is insuarance higher?, see that thing")
```

```python
plt.figure(figsize=(12, 9))
sns.scatterplot(x = 'expenses', y ='age',  sizes=(20, 200), style = 'sex' ,palette = 'flare', data = df)
plt.title("Age and expenses and sex")
```

```python
plt.figure(figsize=(12, 9))

sns.scatterplot(x='expenses', y ='bmi', hue='bmi', data = df, palette = 'flare')
plt.title("Are people with higher bmi more expensive? I'll show you")
```

```python
plt.figure(figsize=(12,9))
sns.histplot(data=df, x = 'age', y = 'expenses',hue = "expenses", legend = False, palette = 'flare')
plt.title("Age may be influence expenses")
```

```python
plt.figure(figsize=(12, 9))
sns.lineplot(x = 'age',y='bmi', hue = 'region', markers = True, data = df, palette = 'flare')
plt.title("I think The old man in the South is more bmi")
plt.show()
```

```python
plt.figure(figsize=(12, 9))
sns.lineplot(x = "smoker", y = "expenses", markers = True,  data = df, palette = 'flare')
plt.title("Don't smoke!")
```

```python
plt.figure(figsize=(12, 9))
sns.lineplot(x='region', y ='expenses' , hue = 'smoker', data = df, legend= True, palette = 'flare')
plt.title("The good region, The bad region")
```
